```python
#1
import pandas as pd
df = pd.read_csv('acc_players-2324F.csv', skiprows=1)
df.head()
```

```python
#2
total_points = df['PTS'].sum()
print("Total points scored by all players:", total_points)

most_minutes_player = df.loc[df['MP'].idxmax()]
print("Player with the most minutes played:", most_minutes_player['Player'], "-", most_minutes_player['MP'], "minutes")

top_5_rebounds = df[['Player', 'TRB']].sort_values(by='TRB', ascending=False).head(5)
print("Top 5 players by total rebounds:")
print(top_5_rebounds)
```

```python
#3
df_filtered = df[df['MP'] > 500]

highest_assists_player = df_filtered.loc[df_filtered['AST'].idxmax()]
print("Player with the highest total assists:", highest_assists_player['Player'], "-", highest_assists_player['AST'], "assists")

top_3_assist_leaders = df_filtered[['Player', 'AST']].sort_values(by='AST', ascending=False).head(3)
print("Top 3 Assist Leaders:")
print(top_3_assist_leaders)

top_3_shot_blockers = df_filtered[['Player', 'BLK']].sort_values(by='BLK', ascending=False).head(3)
print("Top 3 Shot Blockers:")
print(top_3_shot_blockers)
```

```python
#4
school_points = df.groupby('School')['PTS'].sum()
print("Total points scored by each school:")
print(school_points)

school_assists = df.groupby('School')['AST'].sum()
print("Total assists by each school:")
print(school_assists)

top_3_schools_points = school_points.sort_values(ascending=False).head(3)
print("Top 3 schools by total points:")
print(top_3_schools_points)
```

```python
#5
import matplotlib.pyplot as plt
top_5_players_points = df[['Player', 'PTS']].sort_values(by='PTS', ascending=False).head(5)
plt.bar(top_5_players_points['Player'], top_5_players_points['PTS'])
plt.xlabel('Player')
plt.ylabel('Points')
plt.title('Top 5 Players by Total Points')
plt.show()


plt.scatter(df['FG%'], df['WS'])
plt.xlabel('Field Goal Percentage (FG%)')
plt.ylabel('Win Shares (WS)')
plt.title('Correlation between FG% and WS')
plt.show()

correlation = df['FG%'].corr(df['WS'])
print("Correlation between FG% and WS:", correlation)
```
